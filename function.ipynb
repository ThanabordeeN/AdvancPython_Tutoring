# Function Python
ฟังก์ชัน (function) ในการเขียนโปรแกรม คือ บล็อกของคำสั่งซึ่งสามารถเรียกใช้ได้หลายครั้ง โดยมักจะมีชื่อและสามารถรับข้อมูลเข้ามา (parameter) เพื่อประมวลผลและคืนค่า (return) ออกไป

ฟังก์ชันมีคุณสมบัติหลักๆ ดังนี้:

**วัตถุประสงค์ของฟังก์ชัน**
1. ลดความซับซ้อนของโค้ด: ฟังก์ชันช่วยแบ่งโค้ดออกเป็นส่วนๆ ที่จัดการกับงานเฉพาะ ทำให้โค้ดมีความอ่านง่ายและบำรุงรักษาง่ายขึ้น
2. การทำซ้ำ: ฟังก์ชันสามารถเรียกใช้ได้หลายครั้ง โดยไม่ต้องเขียนโค้ดซ้ำๆ
3. การแยกการทำงาน: ฟังก์ชันช่วยแยกการทำงานออกเป็นส่วนๆ ทำให้โค้ดมีความเป็นระเบียบมากขึ้น

**ส่วนประกอบของฟังก์ชัน**
1. ชื่อฟังก์ชัน: เป็นชื่อที่กำหนดให้กับฟังก์ชัน
2. พารามิเตอร์: คือข้อมูลที่ส่งเข้ามาในฟังก์ชันเพื่อประมวลผล
3. คำสั่ง: คือโค้ดที่อยู่ภายในฟังก์ชัน
4. คืนค่า: คือผลลัพธ์ที่ฟังก์ชันส่งออกไปหลังจากประมวลผล

```python
def ตัวอย่างฟังก์ชัน(ชื่อ):
 print("สวัสดี " + ชื่อ)

# เรียกใช้ฟังก์ชัน
ตัวอย่างฟังก์ชัน("จอห์น")
```

## ประเภทของ Function
1. ฟังก์ชันแบบไม่รับค่า
2. ฟังก์ชั่นแบบรับค่าแต่ไม่ Return / คืนค่า
3. ฟังก์ชั่นแบบรับค่าและคืนค่า

## เปรียบเทียบ
ลองนึกภาพว่าเรากำลังทำอาหารเย็น

สมมติว่าเราต้องทำผัดกะเพราไก่บ่อยๆ  แทนที่จะทำทีละขั้นตอนทุกครั้ง  เราสามารถ **เขียนสูตรอาหาร** (function) ไว้เลย  ว่าต้องทำอะไรบ้าง เช่น

1. หั่นไก่
2. สับกระเทียม พริก
3. ผัดกระเทียม พริก
4. ใส่ไก่
5. ปรุงรส
6. ใส่ใบกะเพรา

ทีนี้  เวลาอยากกินผัดกะเพรา  เราก็แค่ **เรียกใช้สูตร** (เรียกใช้ function)  ก็จะได้ผัดกะเพราอร่อยๆ  โดยไม่ต้องทำทีละขั้นตอนใหม่

**Function ในโปรแกรมมิ่ง ก็เหมือนสูตรอาหาร**  มันคือชุดคำสั่งที่เขียนไว้ล่วงหน้า  เพื่อทำงานเฉพาะอย่าง  เช่น

* คำนวณพื้นที่วงกลม
* เรียงลำดับข้อมูล
* ดาวน์โหลดไฟล์จากอินเทอร์เน็ต

**ข้อดีของการใช้ Function:**

* **ไม่ต้องเขียนโค้ดซ้ำๆ:**  ถ้าต้องทำงานเดิมๆ หลายครั้ง  เขียน function ครั้งเดียว แล้วเรียกใช้ได้ตลอด
* **โค้ดอ่านง่าย:**  แบ่งโค้ดเป็นส่วนๆ  เข้าใจง่ายกว่า
* **แก้ไขง่าย:**  ถ้าต้องแก้ไขโค้ด  แก้แค่ใน function  ไม่ต้องแก้หลายที่
* **นำกลับมาใช้ใหม่ได้:**  เขียน function ดีๆ  สามารถนำไปใช้ในโปรแกรมอื่นได้


**สรุป:**  Function คือส่วนประกอบสำคัญในการเขียนโปรแกรม  ช่วยให้โค้ดเป็นระเบียบ  อ่านง่าย  แก้ไขง่าย  และนำกลับมาใช้ใหม่ได้   เหมือนกับการมีสูตรอาหาร  ทำให้เราทำอาหารได้ง่ายขึ้นและรวดเร็วขึ้นนั่นเอง




### ข้อ 1 การเขียนฟังก์ชันแบบไม่รับค่า

````python
def verbose():
    print("This is a verbose function")
    print("It prints out a message")
    print("It does nothing else")
    print("It is not very useful")
    print("It is not very efficient")
````

````python
verbose()

````

### ข้อที่ 2 ฟังก์ชั่นแบบรับค่าแต่ไม่ Return / คืนค่า

````python
def add(a, b):
    print("Adding", a, "and", b)
    print("The result is", a + b)

def subtract(a, b):
    print("Subtracting", b, "from", a)
    print("The result is", a - b)
````

````python
add(3, 4)
print("-"*15)
subtract(7, 2)
````

## Function แบบรับค่า และคืนค่า
บางทีเราอาจจะอยากปรับสูตรผัดกะเพรา ให้อร่อยถูกใจมากขึ้น  เช่น

* **ใส่พริกมากน้อย:**  บางคนชอบเผ็ดมาก  บางคนชอบเผ็ดน้อย
* **ใส่ผักเพิ่ม:**  บางคนชอบใส่ถั่วฝักยาว  บางคนชอบใส่มะเขือ

เราสามารถ **ส่งค่า Argument** ให้กับ function ผัดกะเพรา  เพื่อบอกว่าอยากได้แบบไหน

เช่น

```
ผัดกะเพรา(พริก 5 เม็ด, ใส่ถั่วฝักยาว)  
ผัดกะเพรา(พริก 2 เม็ด, ใส่มะเขือ)
```

**Argument ก็เหมือนวัตถุดิบที่เราส่งให้ function**  เพื่อให้ function ทำงานได้ตามที่เราต้องการ

**ตัวอย่างในโปรแกรมมิ่ง:**

```python
def คำนวณพื้นที่วงกลม(รัศมี):
  พื้นที่ = 3.14 * รัศมี * รัศมี
  return พื้นที่

พื้นที่1 = คำนวณพื้นที่วงกลม(5) # ส่งค่ารัศมี 5
พื้นที่2 = คำนวณพื้นที่วงกลม(10) # ส่งค่ารัศมี 10
```

ในตัวอย่างนี้  `รัศมี`  คือ Argument  ที่เราส่งให้ function `คำนวณพื้นที่วงกลม`   function จะใช้ค่ารัศมีที่เราส่งให้  ไปคำนวณพื้นที่วงกลม


**สรุป:**

การส่งค่า Argument  ทำให้เราสามารถปรับแต่งการทำงานของ function ได้  เหมือนกับการใส่ส่วนผสมต่างๆ ในสูตรอาหาร  เพื่อให้ได้รสชาติที่เราต้องการ


### ข้อ 2 การเขียนฟังก์ชันแบบรับค่าตัวแปร แต่ไม่คืนค่า

````python
def radius(radius_value):
    area = 3.14 * radius_value ** 2
    return area

area_1 = radius(5)
area_2 = radius(7)

print(area_1)
print(area_2)
````

### ข้อ 3 ฟังก์ชั่นแบบรับค่าและคืนค่า

````python
def kitchen(item1, item2):
    return item1 + " and " + item2
````

````python
print(kitchen("pasta", "sauce"))
````

````python
def calculator(a,b,operation):
    if operation == "add":
        return a + b
    elif operation == "subtract":
        return a - b
    else:
        return "I don't know what you want me to do"
````

````python
print(calculator(3, 4, "add"))
````
